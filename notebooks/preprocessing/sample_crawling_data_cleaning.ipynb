```python
import pandas as pd

# 1. CSV 파일 불러오기
df = pd.read_csv('../../data/external/sample_place_restaurant_data.csv')

# 2. 컬럼 순서 지정 (추천 순서)
new_order = [
    '음식점_이름', '카테고리', '주소', '음식점_사진', 
    '영업시간', '전화번호', '음식점_태그', '리뷰', '메뉴_정보'
]

# 3. 컬럼 순서 변경 (없는 컬럼은 제외)
df = df[[col for col in new_order if col in df.columns]]
```

```python
print(df_filtered['음식점_태그'].dtype)
print(df_filtered['리뷰'].dtype)
print(df_filtered['메뉴_정보'].dtype)
```

```python
# 빈 딕셔너리 / 빈 리스트인 행 찾기
mask = (
    (df_filtered['음식점_태그'].apply(lambda x: str(x) == '{}')) &
    (df_filtered['리뷰'].apply(lambda x: str(x) == '[]')) &
    (df_filtered['메뉴_정보'].apply(lambda x: str(x) == '[]'))
) | df_filtered['주소'].apply(lambda x: str(x) == '정보 없음')
df_empty = df_filtered[mask]
df_empty
```

```python
df_filtered_cleaned = df_filtered[~mask]
```

```python
df_filtered_cleaned[df_filtered_cleaned.isnull().any(axis=1)]
```

```python
def replace_nan_with_empty_list(x):
    if pd.isna(x):
        return []
    else:
        return x

df_filtered_cleaned['메뉴_정보'] = df_filtered_cleaned['메뉴_정보'].apply(replace_nan_with_empty_list)
```

```python
# 정제된 데이터 저장
df_filtered_cleaned.to_csv('../../data/preprocessed/sample_crawling_data_cleaned.csv', index=False, encoding='utf-8-sig')
```
